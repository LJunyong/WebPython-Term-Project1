## 1. 주제 및 주제 선정 이유  
  
주제: 코로나-19 바이러스의 조건에 따른 확진자,사망자 수 분석 

주제 선정 이유:  코로나바이러스감염증-19(코로나19, 영어: coronavirus disease 2019, COVID-19)는 2019년 11월부터 중국에서 최초 보고된 후 확산되기 시작해 현재(2024.05.10)까지 전 세계에서 지속되고 있는 범유행전염병이며 수많은 확진자와 사망자를 기록했다.  
이번 데이터 분석을 통해 코로나19 확진율과 사망률에 미치는 조건을 분석하여 코로나19 바이러스의 특성을 규명하고자 한다.  
특히 코로나 백신이 개발되었지만, 백신의 접종 시작 이후 백신의 효과 자체에 대한 의구심이 지속적으로 제기되고 있는 상황이다. 코로나19 백신이 보급된 후 생긴 확진자의 수와 사망자의 수를 데이터를 활용하여 분석함으로써 코로나19 백신의 효과를 입증하기로 하였다. 코로나19 백신의 효과에 대해 분석함으로써, 백신의 효과에 대한 신뢰를 높일 수 있기 때문이다. 추가적으로 코로나19의 확진율을 인구밀도와 연관지어 분석하여 코로나19의 특성을 규명해보기로 하였다.

## 2. 가설 정의 

1. 인구 밀도가 높은 지역은 낮은 지역에 비해 확진율이 높을 것이다.  
* 코로나19의 확진율을 인구밀도와 연관지어 분석하여 코로나19의 특성을 규명해보기로 한다.  


2. 백신 접종자 수의 증가는 시간이 지남에 따라 확진자와 사망자 수의 감소로 이어질 것이다.  
* 백신의 질병 예방와 사망률을 낮추는 효과를 백신 접종자 수에 기반하여 분석한다.  

3. 누적 백신 접종률이 높은 지역은 낮은 지역보다 누적 확진률이 낮을 것이다.
* 지역간의 비교,분석으로 백신의 효과 입증을 더욱 명확하게 한다.

### 데이터 가공

```python
df3= df3.loc[422:980,["구분","송파구","은평구"]]
# "구분","송파구","은평구"을 가져오고, 날짜("구분")를 2021.04.21~2022.10.31까지 가져온다.
df3=df3.reset_index()
del df3["index"]
df3.fillna(0,inplace=True)
#결측치는 fillna를 통해 0으로 바꾼다.
df3["구분"]=df3["구분"].astype(str)
df3["구분"]=df3["구분"].str[:-9]
df3[["송파구","은평구"]]=df3[["송파구", "은평구"]].astype(int)
#"구분" 시리즈의 타입을 str로 바꾼 후 슬라이싱 하여 깔끔하게 바꿔준다. 
df3["송파구 확진율(%)"]=df3["송파구"].div(6596.61)
df3["은평구 확진율(%)"]=df3["은평구"].div(4701.85)
#각 지역의 확진율(%)을 구하기위해, df3["송파구"]와 df3["은평구"]를 div를 통해 (지역 인구 수/100)으로 나눈다.
df3["Month"]=df3["구분"].str[:-3]
#달을 나타내는 "Month"시리즈를 만들어준다.
df3=df3.drop(["송파구","은평구"],axis=1)
df3[["송파구 확진율 누적(%)","은평구 확진율 누적(%)"]]=df3[["송파구 확진율(%)","은평구 확진율(%)"]].cumsum()
#"송파구 확진율 누적(%)","은평구 확진율 누적(%)"시리즈를 만들기 위해 각 지역의 확진율(%)을 누적하여 더한다.
df3
```

### 서울시 자치구별 백신접종자 수 현황

```python
df4.head()
```

```python
df4= df4.transpose()
#데이터프레임의 행과 열을 바꾼다.
df4=df4.iloc[1:231,[1,2,15,27]]
# 2021.02.27일~2022.10.31 까지 송파구와 은평구의 데이터를 가져온다.
df4=df4.reset_index()
del df4["index"]
df4["날짜"]=df4[1]
df4["예방접종 차수"]=df4[2]
df4["송파구 접종자 수(누적)"]=df4[27]
df4["은평구 접종자 수(누적)"]=df4[15]
#시리즈의 이름들을 수정한다.
df4=df4.drop([1,2,15,27],axis=1)
df4[["송파구 접종자 수(누적)","은평구 접종자 수(누적)"]]=df4[["송파구 접종자 수(누적)","은평구 접종자 수(누적)"]].astype(str)
df4[["송파구 접종자 수(누적)", "은평구 접종자 수(누적)"]]=df4[["송파구 접종자 수(누적)", "은평구 접종자 수(누적)"]].replace(',' ,'',regex = True)
df4[["송파구 접종자 수(누적)","은평구 접종자 수(누적)"]]=df4[["송파구 접종자 수(누적)","은평구 접종자 수(누적)"]].astype(int)
# 데이터에서 ","를 제거하기 위해서 값을 모두 str로 바꾼 뒤 replace(',' ,'',regex = True)를 사용하였다.
#그리고 문자열 데이터를 다시 int로 바꿔준다.
df4["송파구 접종률 누적(%)"]=df4["송파구 접종자 수(누적)"].div(6596.61)
df4["은평구 접종률 누적(%)"]=df4["은평구 접종자 수(누적)"].div(4701.85)
# div를 통해 기존 데이터를 (지역 인구 수/100)으로 나눠서 "송파구 접종률 누적(%)","은평구 접종률 누적(%)" 시리즈를 만들어준다.
df4
```

위 그래프들은 백신 접종자 수와 확진자 수, 사망자 수의 증가,감소를 분석하기 위해 서울시의 코로나 백신 당일 접종자 수, 당일 확진자 수, 당일 사망자 수 데이터를 추출하여 그래프를 그린 것이다.(각각 그래프 5,6,7). 그래프5은 코로나 항체 지속 기간을 고려하여 1차,2차,3차,4차 백신 접종 모두 그래프에 나타냈다.위의 그래프들은 각 지역의 확진율을 분석한 그래프이다. 코로나 백신 최초 접종일은 2021년 2월 27일이지만, 그래프 5 만 데이터가 4월 21로 시작한다는 것을 고려해야한다. 또한 서울시의 전체 인구에 대한 분석이기 때문에 비율 대신 단위:(명)을 그대로 사용하였다. 

'백신 접종자 수의 증가는 시간이 지남에 따라 확진자 수, 사망자 수의 감소로 이어질 것이다.' 라는 가설에 대한 판단은 거짓이다.  
반례를 확인할 수 있기 때문이다.  확실히 그래프 7에서 백신의 최초접종(02/27)이 시작된 후 2월부터 4월까지 사망자 수의 감소를 확인할 수 있다. 하지만 그래프 5에서 2022년까지 백신 3차 접종이 이루어졌음에도 오히려 그래프 6,7에서  2022년 3,4월에 확진자와 사망자 수가 폭등했다는 것을 확인 할 수 있다. 이 결과에 사회적 거리두기,방역 조치, 변종 바이러스 등  다양한 원인이 개입했을 가능성이 있지만, 백신의 지속기간과 효과에 대해 의심이 든다. 항체의 지속 시간이 짧거나 예방 효과의 부족으로 인해 다른 요인의 효과가 백신보다 우세했음을 알 수 있기 때문이다. 백신의 효과에 대한 보다 정확한 판단을 가설3에서 할 수 있을 것으로 예상한다.
